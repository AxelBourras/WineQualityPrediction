```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
wines = pd.read_csv('../app/static/Wines.csv')
wines.head()
```

```python
print(wines['quality'].value_counts())
```

```python
fig, ax = plt.subplots(11, figsize=(25, 40)) 
x = wines['quality']
sns.barplot(ax=ax[0],x=x,y=wines['fixed acidity'])
sns.barplot(ax=ax[1],x=x,y=wines['volatile acidity'])
sns.barplot(ax=ax[2],x=x,y=wines['citric acid'])
sns.barplot(ax=ax[3],x=x,y=wines['residual sugar'])
sns.barplot(ax=ax[4],x=x,y=wines['chlorides'])
sns.barplot(ax=ax[5],x=x,y=wines['free sulfur dioxide'])
sns.barplot(ax=ax[6],x=x,y=wines['total sulfur dioxide'])
sns.barplot(ax=ax[7],x=x,y=wines['density'])
sns.barplot(ax=ax[8],x=x,y=wines['pH'])
sns.barplot(ax=ax[9],x=x,y=wines['sulphates'])
sns.barplot(ax=ax[10],x=x,y=wines['alcohol'])
```

```python
wines = wines.drop('Id',axis=1)
cor = wines.corr()
sns.heatmap(cor, square = True, cmap="coolwarm",linewidths=.5,annot=True)
```
